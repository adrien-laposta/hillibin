```python
import getdist.plots as gdplt
import matplotlib as mpl
from getdist.mcsamples import loadMCSamples
import numpy as np
%matplotlib inline
```

```python
cosmo = ['H0', 'ombh2', 'omch2', 'logA', 'ns', 'tau']
fg = ['Aradio', 'Adusty', 'AdustTT', 'Asz', 'Acib', 'Aksz', 'Aszxcib']
```

```python
def pars_from_str(p, str):
    
    if str == 'H0':
        return(p.H0)
    if str == 'cosmomc_theta':
        return(p.cosmomc_theta)
    if str == 'ombh2':
        return(p.ombh2)
    if str == 'omch2':
        return(p.omch2)
    if str == 'logA':
        return(p.logA)
    if str == 'ns':
        return(p.ns)
    if str == 'tau':
        return(p.tau)
    if str == 'Alens':
        return(p.Alens)
    if str == 'Aplanck':
        return(p.Aplanck)
    if str == 'c0':
        return(p.c0)
    if str == 'c1':
        return(p.c1)
    if str == 'c3':
        return(p.c3)
    if str == 'c4':
        return(p.c4)
    if str == 'c5':
        return(p.c5)
    if str == 'Aradio':
        return(p.Aradio)
    if str == 'Adusty':
        return(p.Adusty)
    if str == 'AdustTT':
        return(p.AdustTT)
    if str == 'Asz':
        return(p.Asz)
    if str == 'Acib':
        return(p.Acib)
    if str == 'Aksz':
        return(p.Aksz)
    if str == 'Aszxcib':
        return(p.Aszxcib)
```

```python
def plot_chains(datasets, labels, wanted, filled = True, verb=False):
        
    samples = [loadMCSamples(element, settings = {'ignore_rows': 0.5}) for element in datasets]
    
    
    for i, label in enumerate(labels):
        print("R-1 ({}) : {}".format(label, samples[i].getGelmanRubin())) 
        #print(samples[i].getMargeStats())
        #print(samples[i].getLikeStats())
    #if "../chains/TTEETEET_bin_AL/mcmc" in datasets:
        #ind = datasets.index("../chains/TTEETEET_bin_AL/mcmc")
        #samp = samples[ind]
        #p = samp.getParams()
        #chi2 = p.chi2
        #idmin = np.argmin(chi2)
        #for element in cosmo:
            #print("{} : {}".format(element, pars_from_str(p, element)[idmin]))
        #for element in fg:
            #print("{} : {}".format(element, pars_from_str(p, element)[idmin]))
        
        
        
    gdplot = gdplt.get_subplot_plotter(width_inch=12)
    gdplot.settings.legend_fontsize = 20
    gdplot.settings.axes_fontsize = 15
    gdplot.settings.lab_fontsize = 20
    gdplot.settings.legend_loc = 'upper right'
    gdplot.triangle_plot(samples, wanted, filled = filled, legend_labels = labels)
    #mpl.pyplot.savefig("../plots/RTE.pdf", dpi = 300)
    #mpl.pyplot.savefig("../plots/" +  + ".pdf", dpi = 300)
```

# Plik vs Hillipop ( + $\tau$-prior )

```python
datasets = ['../chains/hillipop_check/plik/mcmc', '../chains/hillipop_check/hill_PR3/mcmc']
labels = [r'Plik TT+TE+EE high$\ell$ + $\tau$-prior', r'Hillipop TT+TE+EE [PR3] + $\tau$-prior']
plot_chains(datasets, labels, cosmo, verb=True)
```
